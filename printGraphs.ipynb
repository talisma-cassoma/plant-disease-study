```python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread

def display_images_with_titles(directory):
    # Get the list of image files in the directory
    image_files = [file for file in os.listdir(directory) if file.endswith('.png')]

    # Sort the image files alphabetically
    image_files.sort()

    # Calculate number of rows and columns for the grid
    num_images = len(image_files)
    num_rows = (num_images - 1) // 3 + 1  # Ceiling division
    num_cols = min(3, num_images)

    # Create a figure to display the images
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

    # Iterate over the image files and display them with titles
    for i, image_file in enumerate(image_files):
        # Read the image
        image = imread(os.path.join(directory, image_file))

        # Determine the position in the grid
        row = i // 3
        col = i % 3

        # If there's only one row, axes is 1D
        if num_rows == 1:
            ax = axes[col]
        else:
            ax = axes[row, col]

        # Display the image
        ax.imshow(image)
        ax.axis('off')  # Turn off axis
        ax.set_title(chr(65 + i))  # Set title as A, B, C, ...
        #ax.set_title(chr(65 + i), loc='left', fontsize=12, pad=10)  # Set title as A, B, C, ... on the left side

    # Remove any empty subplots
    for i in range(num_images, num_rows * num_cols):
        if num_rows == 1:
            fig.delaxes(axes[i])
        else:
            fig.delaxes(axes.flatten()[i])

    # Adjust layout and display the figure
    plt.tight_layout()
    plt.show()

# Directory containing the radar chart images
radars_chart_dir = '/content/radars_graphs'

# Display images with titles
display_images_with_titles(radars_chart_dir)
```

```python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread

def display_images_with_titles_vertically(directory):
    # Get the list of image files in the directory
    image_files = [file for file in os.listdir(directory) if file.endswith('.png')]

    # Sort the image files alphabetically
    image_files.sort()

    # Create a figure to display the images
    fig = plt.figure(figsize=(6, 6))

    # Iterate over the image files and display them with titles
    for i, image_file in enumerate(image_files):
        # Read the image
        image = imread(os.path.join(directory, image_file))

        # Display the image
        plt.imshow(image)
        plt.axis('off')  # Turn off axis
        plt.title(chr(65 + i))  # Set title as A, B, C, ...
        plt.show()

# Directory containing the radar chart images
radars_chart_dir = '/content/models_train'

# Display images with titles
display_images_with_titles_vertically(radars_chart_dir)
```
